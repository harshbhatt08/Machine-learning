## Matplotlib eg workflow:

```python
# Import matplotlib and get it ready for plotting in jupyter 
%matplotlib inline
import matplotlib.pyplot as plt

# 1. prepare data
x = [1,2,3,4] 
y = [11,22,33,44]

# 2. Setup plot
fig, ax = plt.subplots(figsize = (10,5)) # width, height

# 3. Plot the data
ax.plot(x,y)
 
# 4. Customize plot 
ax.set(title= "My new plot", 
      xlabel = "X-axis",
      ylabel = "y-axis")

# 5. Save & show (saving the figure)
fig.savefig("images/myplot.png")
```

# Making figures with Numpy arrays

we want:
* Line plot
* Scatter Plot
* Bar Plot
* Histogram
* Subplots

```python
import numpy as np
```

```python
#create some data
x=np.linspace(0,10,100 )
x
```

```python
#  Plot the data
fig, ax = plt.subplots()
ax.plot(x, x**2)
```

```python
# Use same data top make scatterplot
fig, ax = plt.subplots()
ax.scatter(x,np.exp(x));
```

```python
# Another scatter plot 
fig, ax= plt.subplots()
ax.scatter(x,np.sin(x));
```

```python
# make a plot from directories 
butter = {
    "badaam" : 100,
    "peanut" : 80,
    "Kaju"   : 120
}
fig, ax= plt.subplots()
ax.bar(butter.keys(), butter.values())
ax.set(title= "NUT", ylabel = "Rate");
```

```python
fig, ax= plt.subplots()
ax.barh(list(butter.keys()), list(butter.values()));
```

```python
# Make some data for histogram and plot it
x= np.random.randn(1000)
fig, ax= plt.subplots()
plt.hist(x);
```

## Tow options for subplot


```python
# Subplot option 1
fig, ((ax1,ax2), (ax3,ax4)) = plt.subplots(nrows= 2, ncols=2,figsize=(10,5))


# Plot tot each different axis
ax1.plot(x, x/2)
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(butter.keys(), butter.values());
ax4.hist(np.random.randn(1000));
```

```python
# Subplot optrion 2
fig, ax = plt.subplots(nrows =2, ncols= 2, figsize =(10,5))

# Plot to each figgerent  
# Index the ax variable to plot data
ax[0, 0].plot(x, x/2);
ax[0, 1].scatter(np.random.random(10), np.random.random(10));
ax[1, 0].bar(butter.keys(), butter.values());
ax[1, 1].hist(np.random.randn(1000));
```
